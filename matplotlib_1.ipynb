```python
import matplotlib.pyplot as plt
# this is used because chart can be rended fine.
%matplotlib inline 
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
# controlling the vizualization:
```

```python
# to plot simple 2-D chart:
plt.plot(x,y)
```

```python
# for variants go for documentation: 
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x,y,"g+--")
```

```python
# here order doent matter first color and next type:

plt.plot(x,y,"--r")
```

```python
plt.plot(x,y,color = "green",marker = "D",linestyle = '')
plt.xlabel("scales")
plt.ylabel("temperature")
```

```python
# alpha: ->this will basically controls the transparency
#        ->this will range between 0-1
plt.plot(x,y,"g+-.",alpha = 1)
```

```python
plt.plot(x,y,"g+-.",alpha = 0.5)
```

```python
days  =    [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
# in same chart we can plot multiple graphs:
plt.title("Weather_Report")
plt.plot(days,max_t,"g+:",label = "max_temp")
plt.plot(days,min_t,"b+:",label = "min_temp")
plt.plot(days,avg_t,"r+:",label = "avg_temp")
plt.xlabel("days")
plt.ylabel("temperature")


# this describes the label in graph:
# if we want to change the place of this we have to mention it:

# for various locations and extra use documentation:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

plt.legend(loc = "upper right",shadow = True,fontsize = "large")

plt.grid() # this is more specific way to look over things
```

```python
# Bar_chart:
```

```python
company = ["TATA","Google","Msft","Amazon"]
revenue = [200,180,178,169]
profit  = [195,150,132,128] 
```

```python
# this is normal plotting:

plt.plot(company,revenue)
```

```python
# this is bar plotting:

plt.bar(company,revenue, label = "Revenue")

plt.title("Revenue_Analysis")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.legend(loc = "upper right")
```

```python
# if u wanted to identify them with numbers here we go:
import numpy as np
x_pos = np.arange(len(company))
x_pos
```

```python
plt.xticks(x_pos,company)
# -.02 & +0.2 will get seperated and by width we see it properly. 
plt.bar(x_pos - 0.2,revenue,width = 0.4,label = "Revenue")
plt.bar(x_pos+ 0.2,profit,width = 0.4,label = "Profit") # only in this case we can plot more than one information in bar charts.
plt.ylabel("Revenue")
plt.xlabel("company")    
plt.title("Tech Stocks")
plt.legend()
```

```python
# if we want bar_chart from y-axis:
plt.yticks(x_pos,company)
plt.title("Analysis")
plt.barh(x_pos-0.2,revenue,label = "Revenue")
plt.barh(x_pos+0.2,profit,label = "profit")
plt.legend()
```

```python
# Histo_Grams: kind of bar_charts y-axis has frequency of a variable
```

```python
# need of patients name of sugar samples histogram varies with samples:

blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

# rwidth: is relative width of bar
# bins: these are used to arrange default value:10
plt.hist([blood_sugar_men,blood_sugar_women], 
         bins =[80,100,125,150],
         rwidth = 0.95,
         color = ["g","orange"],
        label = ["men","women"])  # histtype = "step"
plt.legend()
plt.xlabel("Sugar Range")
plt.ylabel("no of patients")
plt.title("Sugar Analysis")
```

```python
# horizontal view:

blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

# rwidth: is relative width of bar
# bins: these are used to arrange default value:10
plt.hist([blood_sugar_men,blood_sugar_women], 
         bins =[80,100,125,150],
         rwidth = 0.95,
         color = ["g","orange"],
        label = ["men","women"],
        orientation = "horizontal")  # histtype = "step"
plt.legend()
plt.xlabel("Sugar Range")
plt.ylabel("no of patients")
plt.title("Sugar Analysis")
```

```python
exp_val = [1400,600,300,410,250]
exp_labels = ["HouseRent","Food","Internet Bill","Car","other utilites"]

plt.pie(exp_val,
        labels = exp_labels,
        shadow = True,
        autopct='%2.1f%%',# shows percentages in pie chart -> "%f%%".
        startangle = 45, # rotates pie chart by counterclock wise from x-axis.
        radius = 2, # use when we want bigger circle default value:1
        # explode:comes out of the pie chart Mandatory->exp_labels == explode 
        explode = [0,0.2,0,0,0])
plt.axis("equal") # this gives exactly circle without irregular shapes.
plt.show() # this removes the unwanted text shows above the piechart.

# save pie chart in files:
plt.savefig("piechart.png",
            bbox_inches = "tight", # this will free ur diagram
            pad_inches = 2, # this enlarges ur figure
            transparent = True # this gives transparent fig 
            )
```
