```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10,2)
y = x**2
```

```python
# Basic implementation of graph.
plt.plot(x,y,'b-.',label = "random")
plt.xlabel("even_nums")
plt.ylabel("squares_of_x")
plt.title("linear_graph")
plt.legend()
plt.show()
```

```python
# what if we give the values inside the .plot()?
plt.plot([10,20,30],'g--',label = "linear")
plt.legend()
plt.show()


# if u give values inside function it will consider it defualtly as y-axis values and x-axis values are taken randomly from 0.
```

```python
# markers representation:
x = [1,2,5,7,9]
y = [i**2 for i in x]
# mec[marker edge color]: outside area of the marker.
# mfc[marker face color]: inside area of the marker.
# ms[marker size]:this is used to make ur marker size big.
plt.plot(x,y,marker = 'o',mec = 'r',mfc = 'g',ms = '10')
plt.title("Linear_Graph")
plt.xlabel("x-axis")
plt.ylabel('Y-axis')
plt.show()
```

```python
# lines(Style,Color,Width):
x = [10,20,30,40,50]
y = [i**2 for i in x]
# ls: line style,lw: line width,c = color
plt.plot(x,y,ls = '-.',lw = 2,c = 'g')
plt.show()
```

```python
# draw a graph of squares,cubes of x-values by mentioning it.
x  = [1,2,3,4,5,6,7]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]
plt.plot(x,y1,marker = 'o',mec = 'k',mfc = 'r',ls = '-.',lw = 1.5,c = 'g',label = "Squares")
plt.plot(x,y2,marker = 'o',mec = 'r',mfc = 'k',ls = ':',lw = 1.5,c = 'b' ,label = "Cubes")
plt.title("Natural Numbers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# Grid Representation:
x = [1,2,3,4,5]
y = [30,10,20,40,50]
plt.plot(x,y)
# axis = 'x': it gives only grid lines for x-axis vice versa y if it is both it gives x,y
# ls,lw,c are similar u can give what ever u want
plt.grid(ls = ':',lw = 1,c = 'k')
plt.show()
```

```python
# sub-plot: we can fit no.of graphs in a single figure this is done with the help of subplot()
# arguments: r,c,i r = rows,c = columns,i = index of plot

x  = [0,1,2,3,4]
y1 = [2,3,4,5,6]
y2 = [3,4,5,6,7]

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("First plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Second plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.suptitle("Example")
plt.show()
```

```python
x  = [10,20,23,4,2,21]
y1 = [12,4,32,52,56,22]
y2 = [23,1,5,14,12,19]



plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("First plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("Second plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
# Legend():
x = [1,2,3,4,5]
y = [16,2,12,14,4]
plt.plot(x,y,c = 'g',label = 'Sample_one')
# locations are described by numbers 0-10.
# framealpha: this is 0-1 this gives the transparency of the box. 
# facecolor: the color represnt inside the box of representation.default["white"]
# edgecolor: this is the border color of the box.
# shadow: it is boolean it defaultly False if u turned it into True it gives a shadow appearence.
#fancybox: this is also a boolean if it is false it is in rectangular if it is True then it will be appeared as edges of the box.


plt.legend(loc = "upper left",fancybox = True,edgecolor = "green",facecolor = 'orange',shadow = True,fontsize = 12,framealpha = 0.4)
plt.show()
```

```python
# Scatterplot : this consist of only points no lines are used in this graph.

x = [1,2,3,4,5]
y = [2,3,4,5,6]
z = [3,1,5,2,5]

# default colors of 1-plot-> Blue,2-plot->Orange.
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()
```

```python
# change the color, marker in the scatter plot. and plot the graph.
y = [2,3,4,5,6]
z = [3,1,5,2,5]

# s =Size of marker,alpha = transperency.
plt.scatter(y,z,color = "red",marker = "*",alpha = 0.8,linewidth = 5,s = 50)
plt.show()
```

```python
# plot a scatter plot each point with different color,size.

x = [1,2,3,4,5]
y = [200,400,100,300,500]

# colors should be same as x-values,y-values.
colors = ["green","pink","black","blue","red"]
sizes = [121,105,204,180,200]
plt.scatter(x,y,c = colors,s = sizes)
#color bar is used to describe the amount of color present in graph.
plt.colorbar()
plt.show()
```

```python
# Bar Plot: 
# bar: this represent vertically,barh: it represents vertically.
x = ["reliance","amazon","Tesla","TATA"]
y = [5,10,15,80]

plt.bar(x,y,color = "black",width = 0.4)
plt.show()
```

```python
# this is horizontal bar graph.

x = ["reliance","amazon","Tesla","TATA"]
y = [5,10,15,80]

plt.barh(x,y,color = "black",height = 0.4)
plt.show()
```

```python
# change the color for each bar and plot it.
x = ["A","B","C","D"]
y = [10,20,30,40]
colors = ["red","green","black","pink"]
plt.bar(x,y,color = colors,width = 0.4)
```

# Grouped Bar Chart || Multiple Bar Chart :

```python
years = ["2017","2018","2019","2020"]

cse   = [80,90,108,150]
ece   = [30,46,68,80]
mech  = [60,80,90,105]
civil = [50,70,90,100]

w = 0.2

# this is to increase the size of the figure 
plt.figure(figsize = (10,5))

#LOCATIONS TO PLOT
cse_bar = np.arange(len(years))
ece_bar = [i+w for i in cse_bar]
mech_bar = [i+w for i in ece_bar]
civil_bar = [i+w for i in mech_bar]

#ACTUAL PLOT
print(cse_bar)
plt.bar(cse_bar,cse,width=w,label="cse")
print(ece_bar)
plt.bar(ece_bar,ece,width=w,label="ece")
print(mech_bar)
plt.bar(mech_bar,mech,width=w,label="mech")
print(civil_bar)
plt.bar(civil_bar,civil,width=w,label="civil")

# this will add years to your x-axis.
plt.xticks(cse_bar,years)
plt.ylim(0,200)
plt.title("Placement_UG")
plt.xlabel("years")
plt.ylabel("no.of placements")
plt.legend()
plt.show()
```

## PIE CHARTS

```python
stud_rem = ["Excellent","Good","Average","Poor"]
stud_sco = [15,25,12,9]


#startangle : the angle to be started.
# explode : when u need any part of pie chart out of it u need this.
# shadow : there will be changes in pie chart when u apply it.
#autopct : this shows the percentage of each portion in piechart.

plt.figure(figsize = (7,10))
plt.pie(stud_sco,labels = stud_rem,
         startangle = 90,
         explode = [0.2,0,0,0],
         shadow = True,
         colors = ["red","green","blue","black"],
         autopct = "%2.1f%%")
#pie chart is in normal size plt legend will overlap so,increase fig size.
plt.legend(title = "Performances")
plt.show()
```

# HISTO GRAMS

```python
x = np.array([10,25,38,45,49,50,57,62,69,70,78,80,89,92,95,100])
grade_intervals  = np.array([0,40,70,100])
plt.title("Grades_Pie-Chart")

# parameteres:
# histtype : step,bar
# rwidth : this is going to be the width of each bar
# facecolor : this shows the color of your histogram.

plt.hist(x,grade_intervals,histtype = "bar",rwidth = 0.4,facecolor = "green")
plt.xticks([0,40,70,100])
plt.xlabel("percentage")
plt.ylabel("no.of.students")
plt.show()
```
